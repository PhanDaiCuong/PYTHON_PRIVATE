```python
from math import *
from functools import cmp_to_key

def cmp(a, b):
    return b.getNamSinh() - a.getNamSinh()
    
class Person:
    def __init__(self, name, namsinh):
        self.name = name
        self.namsinh = namsinh
    def getNamSinh(self):
        return self.namsinh
        
    def describe(self):
        print('Name: ' + self.name + " - yob: " + str(self.namsinh), end = '')
        
class Teacher(Person):
    def __init__(self, name, namsinh, subject):
        super().__init__(name, namsinh)
        self.subject = subject
    def describe(self):
        print('Teacher - ', end = '')
        super().describe()
        print(' - subject:  ' + self.subject)
        
class Student(Person):
    def __init__(self, name, namsinh, grade):
        super().__init__(name, namsinh)
        self.grade = grade
    def describe(self):
        print('Student - ', end = '')
        super().describe()
        print(' - Grade: ' + self.grade)
class Doctor(Person):
    def __init__(self, name, namsinh, specialist):
        super().__init__(name, namsinh)
        self.specialist = specialist
    def describe(self):
        print("Doctor - ", end = '')
        super().describe()
        print(' - Specialist: ' + self.specialist)
class Ward:
    def __init__(self, name):
        self.name = name
        self.people = []
    def addPerson(self, person):
        self.people.append(person)
    def describe(self):
        print('Ward Name: ' + self.name)
        for x in self.people:
            x.describe()
    # Ham dem so luong Doctor trong Ward
    def countDoctor(self):
        cnt = 0
        for x in self.people:
            if isinstance(x, Doctor):
                cnt += 1
        return cnt
    #Ham sort moi nguoi trong Word theo tuoi tho tang dan
    def sortAge(self):
        self.people.sort(key = cmp_to_key(cmp))
    #Ham tinh trung binh nam sinh cua cac teacher
    def aveTeacherYearOfBirth(self):
        cnt = 0
        Sum = 0
        for x in self.people:
            if isinstance(x, Teacher):
                cnt += 1
                Sum += x.getNamSinh()
        return Sum // cnt
        
if __name__ == '__main__':
    #1
    student1 = Student("studentA", 2010, "7")
    student1.describe()
    teacher1 = Teacher("teacherA",1969,"Math")
    teacher1.describe()
    
    doctor1 = Doctor("doctorA", 1945,"Endocrinologists")
    doctor1.describe()
    #2
    print()
    teacher2 = Teacher("teacherB", 1995, "History")
    doctor2 = Doctor("doctorB", 1975, "Cardiologists")
    ward1 = Ward("Ward1")
    ward1.addPerson(student1)
    ward1.addPerson(teacher1)
    ward1.addPerson(teacher2)
    ward1.addPerson(doctor1)
    ward1.addPerson(doctor2)
    ward1.describe()
    #3
    print('So luong Doctor trong Ward la: ', ward1.countDoctor())
    #4
    print()
    print('Thong tin cua Ward sau khi sap xep theo tuoi tang dan la: ')
    ward1.sortAge()
    print()
    ward1.describe()
    #5
    print()
    print('Nam sinh trung binh cua cac Teacher la: ', ward1.aveTeacherYearOfBirth())
```
